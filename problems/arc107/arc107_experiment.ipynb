```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def mex(A):
    ret = np.copy(A)
    for i in range(1, A.shape[0]):
        for j in range(1, A.shape[1]):
            x = ret[i-1][j]
            y = ret[i][j-1]
            if x and y:
                ret[i][j] = 0
            elif x==1 or y==1:
                ret[i][j] = 2
            else:
                ret[i][j] = 1
    return ret
```

```python
N = 20
Ai = np.random.randint(0, 3, N)
Aj = np.random.randint(0, 3, N)
A = np.zeros((N,N))
for i in range(N):
    A[i][0] = Ai[i]
    A[0][i] = Aj[i]
A_mex = mex(A)
sns.heatmap(A_mex)
```
