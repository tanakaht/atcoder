```python
import requests
from bs4 import BeautifulSoup
import sys
import time
import json

loginconfig = dict(username='hte', password='qutvHiro2279')
# パラメータセット
# パラメータセット
problem = "typical90_bm"
lang = "pypy3"
LOGIN_URL = "https://atcoder.jp/login"
SUBMIT_URL = f"https://atcoder.jp/contests/{problem.split('_')[0]}/submit"
PROBLEM_URL = f"https://atcoder.jp/contests/{problem.split('_')[0]}/tasks/{problem}"
PYTHON_ID = 4006
PYPY_ID = 4047
LANG_ID = PYPY_ID
filename = f"/Users/ht/googledrive/projects/atcoder/problems/{problem.split('_')[0]}/{problem}.py"
print(filename)
with open(filename, 'r') as f:
  source_code = f.read()

# セッション開始
session = requests.session()

# login
# csrf_token取得
r = session.get(LOGIN_URL)
s = BeautifulSoup(r.text, 'lxml')
csrf_token = s.find(attrs={'name': 'csrf_token'}).get('value')
login_info = {
    "csrf_token": csrf_token,
    "username": loginconfig['username'],
    "password": loginconfig['password']
}
result = session.post(LOGIN_URL, data=login_info)
result.raise_for_status()
if result.status_code==200:
  print("log in!")
else:
  print("failed...")


# 提出
cnt = 0
# ページできるま待機
while True:
    try:
        html = session.get(PROBLEM_URL)
        html.raise_for_status()
        break
    except:
        time.sleep(1)
        cnt += 1
        if cnt%100==0:
            print(f'waiting...({cnt})')
        continue

html = session.get(SUBMIT_URL)
html.raise_for_status()
soup = BeautifulSoup(html.text, 'lxml')
csrf_token = soup.find(attrs={'name': 'csrf_token'}).get('value')

submit_info = {
    "data.TaskScreenName": problem,
    "csrf_token": csrf_token,
    "data.LanguageId": LANG_ID,
    "sourceCode": source_code
}

result = session.post(SUBMIT_URL, data=submit_info)
result.raise_for_status()

if result.status_code == 200:
    print("Submitted!")
    break
else:
    time.sleep(1)
```
