```python
def state2num(edges):
    ret = 0
    for i in range(len(edges)):
        ret += edges[i]
        ret = ret << 5
    ret = ret >> 5
    return ret

def num2state(num):
    ret = []
    for i in range(5):
        ret.append(num%(1<<5))
        num = num >> 5
    return ret[::-1]

edges = [1<<i for i in range(4)] + [0]
print(edges)
n = state2num(edges)
print(n)
print(num2state(n))
```
